<h1 style="text-align: center;display:table; margin:auto;padding:5px;font-size:45px;background-color:#76b5c5;color:black">Dr. Fix-It To Fix The Mess </h1>

<div style="width:100%;text-align: center;"> <img align=middle src="https://c8.alamy.com/comp/TAWMM0/happy-hand-drawn-stick-figure-character-in-hurry-to-help-to-fix-something-TAWMM0.jpg" alt="Fix-it" style="height:300px;margin-top:3rem;"> </div>

<h3 style=";display:table; font-size:20px;background-color:#ff6666;color:black">Steps</h3>



1. Fix the Rating column

2. Remove "," from Cost column

3. Remove the special charachters from the name column

4. Drop the duplicated rows

# 

```python
df = df[(df['rating'] != 'NEW') & (df['rating'] != '-') ]
df['rating'] = df['rating'].apply(lambda x: x.replace("/5",""))
df['rating'] = df['rating'].apply(lambda x: x.replace(" ",""))
df['rating'] = df['rating'].astype('float')
```

```python
df['cost_for_two'] = df['cost_for_two'].apply(lambda x:x.replace(",","")).astype('int')
```

```python
df['name'] = df['name'].str.replace('[Ãx][^A-Za-z]+','',regex=True)
df['name'] = df['name'].str.replace('Caf-|Caf ','Cafe ',regex=True)
```

```python
df = df.drop_duplicates().reset_index()
```

<h5 style="display:table;font-size:30px;background-color:#ff99ff;color:black">The Questions We Need To Ask The Dataset [PART 1]</h5>

1. How many restaurants take online orders ?

2. How many restaurants allow table booking/reservation ?

3. Which 5 locations have highest and lowest rated restaurants ?

4. Top 50 Highest priced retsaurants vs their ratings ?







```python
fig,ax=plt.subplots(figsize=(18,10))


plt.subplot(2,1,1)
colors = ['#66b3ff','#66ffb3']
plt.pie(x=df['online_order'].value_counts(),labels=['Yes','No'],labeldistance=0.85,autopct='%1.1f%%',pctdistance=0.5,colors=colors)
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('How Many Restaurants Take Online Order ?',fontsize=15)
fig.tight_layout()


plt.subplot(2,1,2)
colors = ['#9966ff','#ff80bf']
plt.pie(x=df['book_table'].value_counts(),labels=['No','Yes'],labeldistance=0.85,autopct='%1.1f%%',pctdistance=0.5,colors=colors)
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)


plt.title('How Many Restaurants Allow Table Booking?',fontsize=15)
fig.tight_layout()
```

<h3 style="display:table;background-color:#80ff80;color:black">Observations</h3>

Overwhelming majority of restaurants do not allow table booking

Most restaurants take online orders



# 

```python
pop_loc = df.groupby('location')['rating'].mean().reset_index()


fig = plt.figure(figsize=(15,8))
fig.subplots_adjust(hspace=2.0)
plt.subplot(2,1,1)

bars = sns.barplot(x='location',y='rating',data=pop_loc.sort_values(by='rating',ascending=False).head(5))
plt.title('5 locations With Highest Rated Restaurants',fontsize=13)



plt.subplot(2,1,2)
bars = sns.barplot(x='location',y='rating',data=pop_loc.sort_values(by='rating',ascending=True).head(5))
plt.title('5 locations With Lowest Rated Restaurants',fontsize=13)

fig.tight_layout()
```

<h3 style="display:table;background-color:#80ff80;color:black">Observations</h3>

Lavelle Road has the highest rated restaurants (over mean)


Old madras road has the lowest rated restaurants

```python
price_rest = df.drop_duplicates(subset=['name']).sort_values(by=['cost_for_two'],ascending=False).head(50).reset_index()[['name','rating','cost_for_two']]

import plotly.express as px

fig = px.scatter(price_rest, x="cost_for_two", y="rating",hover_data=price_rest ,size_max=200, color="rating")
fig.update_layout(title_text='Cost vs Rating', title_x=0.5)
print('Hover over the data to get the complete info')
fig.update_layout({
    'plot_bgcolor': 'rgba(0, 0, 0, 0)',
    'paper_bgcolor': 'rgba(0, 0, 0, 0)'})

fig.show()
```

<h3 style="display:table;background-color:#80ff80;color:black">Observations</h3>

"The Leela palace" is the costliest restaurant (priced at 6000.rs for two people)

"Grill 99 the ritz" has the lowest rating among the 50 costliest restaurants (3.5 mean rating)

"Rim naam - The Oberoi" has the Highest rating among the 50 restaurants (4.6 mean rating)

```python
resttype = df.assign(rest_type=df['type'].str.split(',')).explode('rest_type')
type_res = resttype.groupby('rest_type')['name'].count().reset_index()
fig = plt.figure(figsize=(15,15))
ax = sns.barplot(x='name',y='rest_type',data=type_res.sort_values(by=['name'],ascending=False).head(40),orient='h',palette='rocket_r')
ax.set(xlabel='Count', ylabel='Restaurant Type')

plt.title('Types Of Restaurants',fontsize=15)
fig.tight_layout()
```

<h3 style="display:table;background-color:#80ff80;color:black">Observations</h3>

Majority of the restaurants are "Quick bites" type followed by "casual dining" restaurants

```python
price_rest = df.drop_duplicates(subset=['name']).sort_values(by=['cost_for_two'],ascending=False).reset_index()[['name','rating','cost_for_two']]

import plotly.express as px

fig = px.scatter(price_rest, x="cost_for_two", y="rating",marginal_x="histogram", marginal_y="histogram",hover_data=price_rest ,size_max=200, color="rating")
fig.update_layout(title_text='Cost vs Rating', title_x=0.5)
print('Hover over the data to get the complete info')
fig.update_layout({
    'plot_bgcolor': 'rgba(0, 0, 0, 0)',
    'paper_bgcolor': 'rgba(0, 0, 0, 0)'})
fig.show()
```

Cost is left skewed (as expected) i.e there are more Cheaper restaurants than Costlier


Ratings follow uniform distribution and majority of ratings are between 4.2 and 3

"Srinidhi sagar" has the lowest price for two (40.rs) with average rating of 3.9

There are 6 restaurants with 4.9 rating (Highest)


Lowest rated Restaurant is "Alibi"

```python
cuistype = df.assign(cuis_type=df['cuisines'].str.split(',')).explode('cuis_type')
cuistype['cuis_type'] = cuistype['cuis_type'].apply(lambda x:str(x).replace(" ",""))
cuistype['cuis_type'].unique()


type_cuis = cuistype.groupby('cuis_type')['name'].count().reset_index().sort_values(by='name',ascending=False)
fig = plt.figure(figsize=(15,15))
ax =sns.barplot(x='name',y='cuis_type',data=type_cuis.sort_values(by=['name'],ascending=False).head(50),orient='h',palette='rocket')
ax.set(xlabel='Count', ylabel='Cuisine Type')
plt.title('Types Of Cuisines',fontsize=13)
fig.tight_layout()
```

<h3 style="display:table;background-color:#80ff80;color:black">Observations</h3>


No surprises and north indian cuisine is mosty served in restaurants, closely followed by chinese and then south indian
